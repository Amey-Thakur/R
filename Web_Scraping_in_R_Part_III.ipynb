```r
# Load the libraries
library(dplyr)
library(rvest)
library(xml2)
```

```r
# Read the link
#link = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating"

movies = data.frame()

for (page_result in seq(from = 1, to = 250, by = 50 )) {
    link = paste0("https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=", page_result ,"&ref_=adv_nxt")
    page = read_html(link)
    
    name = page %>% html_nodes(".lister-item-header a") %>% html_text()
    director = page %>% html_nodes(".text-muted+ p a:nth-child(1)") %>% html_text()
    year = page %>% html_nodes(".text-muted.unbold") %>% html_text()
    rating = page %>% html_nodes(".ratings-imdb-rating strong") %>% html_text()
    synopsis = page %>% html_nodes(".ratings-bar+ .text-muted") %>% html_text()
    
    movies = rbind(movies, data.frame(name, director, year, rating, synopsis, stringsAsFactors = FALSE))
    
    print(movies)
    
    print(paste("Page:", page_result))
}

write.csv(movies, "movies.csv")
```
