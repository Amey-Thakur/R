```r
#use of piping operator %>%

df <- mtcars
df
```

```r
#nesting

result <- arrange(sample_n(filter(df,mpg>20), size=5), desc(mpg))
result
```

```r
#multiple assignment

a<- filter(df, mpg>20)
b<- sample_n(a, size=5)
result <- arrange(b,desc(mpg))
result
```

```r
#same using pipe operator
#syntax - data %>% op1 %>% op2 %>% op3

result <- df %>% filter(mpg>20) %>% sample_n(size=10) %>% arrange(desc(mpg))
result
```

### Data Manipulation using tidyr


```r
install.packages('tidyr')
library(tidyr)
```

```r
n = 10

wide <- data.frame(
    ID = c(1:n),
    Face.1 = c(411,723,325,456,759,612,709,513,527,379),
    Face.2 = c(123,300,400,500,600,645,789,906,416,567),
    Face.3 = c(1457,1000,569,896,956,2345,780,599,1023,673))

head(wide)
```

```r
#gather() - to reshape the data

long <- wide %>% gather(Face, ResponseTime, Face.1:Face.3)
long
```

```r
#separate() - splits a single column into multiple columns

long_separate <- long %>% separate(Face, c('Target', 'Number'))
long_separate
```

```r
#unite() - combines multipl columns into a single column
long_unite <- long_separate %>% unite(Face, Target, Number, sep = '.')
long_unite
```

```r
#spread - takes two columns and spreads into multiple columns
back_to_wide <- long_unite %>% spread(Face, ResponseTime)
back_to_wide
```
